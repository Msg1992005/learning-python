```python
import pandas as pd
from sklearn.datasets import load_digits
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
data = load_digits()
dir(data)
```

```python
datas = data.data
images = data.images
numbers = data.target
```

```python
x_test ,x_train ,y_test , y_train = train_test_split(datas,numbers)
```

```python
from math import isclose
    
close_scores = {}
for i in range(1,100):
    model = SVC(C=i)
    model.fit(x_train,y_train)
    score = model.score(x_test,y_test)
    if(isclose(score , 0.98,abs_tol=0.01)):
        close_scores[i] = score
print(close_scores) 
```

```python
biger_num = 0
c = 0
for m in close_scores:
    if close_scores[m]> biger_num:
        biger_num = close_scores[m]
        c = m
print(biger_num ,c)
```
