```python
import pandas as pd
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
# from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
x = iris.data
y = iris.target
```

```python
cross_val_score(LogisticRegression(),x,y)
```

```python
cross_val_score(SVC(C=100),x,y)
```

```python
cross_val_score(RandomForestClassifier(n_estimators=3),x,y)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(cross_val_predict(SVC(C=100),x,y),y)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sn
plt.figure(figsize=(3,3))
sn.heatmap(confusion_matrix(cross_val_predict(SVC(C=100),x,y),y),annot=True)
```

```python
plt.figure(figsize=(3,3))
sn.heatmap(confusion_matrix(cross_val_predict(RandomForestClassifier(n_estimators=3),x,y),y),annot=True)
```

```python
plt.figure(figsize=(3,3))
sn.heatmap(confusion_matrix(cross_val_predict(LogisticRegression(),x,y),y),annot=True)
```
